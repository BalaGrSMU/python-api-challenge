#### Latitude vs. Temperature Plot

```python
# define null hypothesis and run regression analysis to test the hypothesis
# null hypothesis - there is no relationship between temperature and latitude
# x, y = latitude, temperature
slope_lat_tem, intercept_lat_tem,r_value_lat_tem,p_value_lat_tem,std_err_lat_tem = st.linregress(lat_data,tmp_data)
print(f"The correlation between Latitude and Temperature is: {r_value_lat_tem}")
print(f"The pvalue is {p_value_lat_tem}")
print(f"The standard error is {std_err_lat_tem}")
if p_value_lat_tem<=0.05:
    print("We reject the null hypothesis that there is no relationship between Latitude and Temperature and \n"
           "accept alternative hypothesis that Latitude may impact Temperature")
else:
    print("We do not reject the null hypothesis and additional data analysis required")
```

```python
# Build Scatter Plot of Temp (F) vs. Latitude

plt.scatter(city_final["latitude"], city_final["temperature"], marker="o")

# Incorporate the other graph properties
plt.title("City Latitude vs. Temperature (F)")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitutde")
plt.grid(True)

# Save the figure
plt.savefig("Images/Latitude_vs_Temp.png")

# Show plot
plt.show()
```
